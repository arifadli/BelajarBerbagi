```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameter distribusi Normal
mu_normal = 0
sigma_normal = 1

# Membuat data acak yang mengikuti distribusi Normal
data_normal = np.random.normal(mu_normal, sigma_normal, 1000)

# Berbagai lebar interval yang akan diuji
bin_widths = [0.1, 0.5, 1.0, 2.0]

# Menghitung jumlah subplot yang dibutuhkan berdasarkan jumlah lebar interval
num_subplots = len(bin_widths)

# Mengatur ukuran gambar berdasarkan jumlah subplot
plt.figure(figsize=(12, 4 * num_subplots))

# Loop untuk membuat masing-masing subplot
for i, bin_width in enumerate(bin_widths, 1):
    # Menghitung jumlah bin berdasarkan lebar interval
    num_bins = int((data_normal.max() - data_normal.min()) / bin_width)

    # Membuat histogram
    plt.subplot(num_subplots, 2, 2 * i - 1)
    plt.hist(data_normal, bins=num_bins, density=True, alpha=0.6, color='g', label='Histogram')
    plt.xlabel('Nilai')
    plt.ylabel('Frekuensi')
    plt.title(f'Histogram dengan Lebar Interval {bin_width}')

    # Menghitung nilai PDF distribusi Normal
    x = np.linspace(-5, 5, 400)
    pdf_normal = norm.pdf(x, mu_normal, sigma_normal)

    # Plotting kurva PDF
    plt.subplot(num_subplots, 2, 2 * i)
    plt.plot(x, pdf_normal, 'r', label='PDF (Normal)')
    plt.xlabel('Nilai')
    plt.ylabel('Densitas Probabilitas')
    plt.title(f'PDF (Normal)')

# Menampilkan layout yang lebih rapi
plt.tight_layout()
plt.show()
```

**Normalisasi Histogram**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameter distribusi Normal
mu_normal = 0
sigma_normal = 1

# Membuat data acak yang mengikuti distribusi Normal
data_normal = np.random.normal(mu_normal, sigma_normal, 1000)

# Berbagai lebar interval yang akan diuji
bin_widths = [0.1, 0.5, 1.0, 2.0]

# Menghitung jumlah subplot yang dibutuhkan berdasarkan jumlah lebar interval
num_subplots = len(bin_widths)

# Mengatur ukuran gambar berdasarkan jumlah subplot
plt.figure(figsize=(12, 4 * num_subplots))

# Loop untuk membuat masing-masing subplot
for i, bin_width in enumerate(bin_widths, 1):
    # Menghitung jumlah bin berdasarkan lebar interval
    num_bins = int((data_normal.max() - data_normal.min()) / bin_width)

    # Membuat histogram yang dinormalisasi menggunakan frekuensi relatif
    plt.subplot(num_subplots, 2, 2 * i - 1)
    plt.hist(data_normal, bins=num_bins, density=False, alpha=0.6, color='g', label='Histogram')
    plt.xlabel('Nilai')
    plt.ylabel('Frekuensi')
    plt.title(f'Histogram dengan Lebar Interval {bin_width}')

    # Normalisasi histogram dengan frekuensi relatif
    plt.hist(data_normal, bins=num_bins, density=True, alpha=0.6, color='g', label='Histogram (Normalized)')

    # Menghitung nilai PDF distribusi Normal
    x = np.linspace(-5, 5, 400)
    pdf_normal = norm.pdf(x, mu_normal, sigma_normal)

    # Plotting kurva PDF
    plt.subplot(num_subplots, 2, 2 * i)
    plt.plot(x, pdf_normal, 'r', label='PDF (Normal)')
    plt.xlabel('Nilai')
    plt.ylabel('Densitas Probabilitas')
    plt.title(f'PDF (Normal)')

# Menampilkan layout yang lebih rapi
plt.tight_layout()
plt.show()
```
